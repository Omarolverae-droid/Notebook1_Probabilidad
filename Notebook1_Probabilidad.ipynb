# **Introducción Probabilidad** 

## ***Experimentos Aleatorios y Eventos***

Un experimento aleatorio es un experimento $\mathcal{E}$ que reproducido en condiciones idénticas puede conducir a varios resultados, en el cual no se puede predecir el resultado.

El espacio de todos los posibles resultados de un experimento aleatorio es llamado espacio de estados <span style="color:red;">espacio de estados</span> y es denotado por un $\Omega$. Un posible resultado del experimento aleatorio se denota por $\omega$ , donde $\omega$ $\mathcal{E}$ $\Omega$.

$\color{red}{\blacktriangleright}$ El lanzamiento de una moneda: $\Omega$ = {águila,sol}.

$\color{red}{\blacktriangleright}$ El lanzamiendo de un dado: $\Omega$ = {1,2,3,4,5,6}.

$\color{red}{\blacktriangleright}$ El lanzamiendo de un dardo en un circulo de 30cm de diámetro: $\Omega = \{(x,y) : \sqrt{x^2 + y^2} \leq 15\}.$
            
$\color{red}{\blacktriangleright}$ Romeo espera a Julieta, el cual va desde 1 minuto a 1 hora: $\Omega$ = [0,1]

$\color{red}{\blacktriangleright}$ Cuando estdiamos el tiempo de vida de una bacteria: $\Omega$ = $[0, \infty)$




Un <span style="color:red;">evento aleatorio</span> (asociado al experimento $\mathcal{E}$ ) es un subconjunto de $\Omega$, donde se puede decir si el experiemnto se realiza o no. Un evento aleatorio $A$ es un subconjunto de $\Omega$.

$\color{red}{\blacktriangleright}$ A: Se obtiene sol al lanzar una moneda.

$\color{red}{\blacktriangleright}$ A: Se obtiene un 3 al lanzar un dado.

$\color{red}{\blacktriangleright}$ Julieta se retrasa menos de un cuarto de hora: $A$ = $[0, 1/4]$

Así, un evento aleatorio esta representado por un conjunto de resultados para los cuales este se realiza.

$P(\Omega)$ = Conjunto de todos los subconjuntos de $\Omega$

- Evento cierto: $\Omega$

- Evento imposible: $\emptyset$ (Conjunto vacío)

- Unión (A o B): $A \cup B$ 

- Intersección (A y B al mismo tiempo): $A \cap B$ 

- Complemento (Todo lo que NO es A): $A^c$ o $A'$ 







## ***Espacios de probabilidad***

Un espacio de probabilidad es una terna $(\Omega, \mathcal{F}, \mathbb{P})$, donde $\Omega$ es un conjunto no vacío, $\mathcal{F}$ es la familia de subconjuntos de $\Omega$ y $\mathbb{P}$ es una medida de probabilidad, la cual le asigna un número no negativo a un evento, está ultima es una función $\mathbb{P}$ : $\mathcal{F}$ $\rightarrow [0,1]$ que verifica los siguientes axiomas:

1. Si $A$ $\mathcal{E}$ $\mathcal{F}$, entonces 0 $\leq \mathbb{P}(A) \leq 1 $. (No negativa)
2. La probabilidad del espacio de estados $\Omega$ es igual a 1, es decir, $\mathbb{P} (\Omega) = 1$. Normalización
3. Si $(A_n)_{n \geq 1}$ es una sucesión finita o infinita numerable de elementos de $\mathcal{F}$, que son mutuamente excluyentes dos a dos (es decir, $A_i \cap A_j = \emptyset$ para $i \neq j)$ entonces

$$\mathbb{P} \left( \bigcup_{n=1}^{\infty} A_n \right) = \sum_{n=1}^{\infty} \mathbb{P}(A_n)$$



### Propiedades:

1. $\mathbb{P}(\emptyset) = 0$
2. $\mathbb{P}(A^c) = 1 - \mathbb{P}(A)$
3. $A \subset B \Rightarrow \mathbb{P}(B \setminus A) = \mathbb{P}(B) - \mathbb{P}(A)$
4. $\mathbb{P}(A \cup B) = \mathbb{P}(A) + \mathbb{P}(B) - \mathbb{P}(A \cap B)$

    $
\begin{align*}
\mathbb{P}(A \cup B \cup C) =\;& \mathbb{P}(A) + \mathbb{P}(B) + \mathbb{P}(C) \\
& - \mathbb{P}(A \cap B) - \mathbb{P}(A \cap C) - \mathbb{P}(B \cap C) \\
& + \mathbb{P}(A \cap B \cap C)
\end{align*}
$

# ***Modelo clasico Laplaciano***

La definición clásica de probabilidad fue introducida en el año 1812 por Pierre-Simon Laplace en el concepto de equiprobabilidad como una propiedad objetiva de los eventos asociados a un experimento aleatorio.

Supongamos que $\Omega$ = ${(\omega_1, \omega_2, ...., \omega_n)}$ es un conjunto finito no vacío y sea $\mathcal{F}$ = $\mathbb{P}(\Omega)$, los elementos ${\omega_i}$ tienen la misma probabilidad $1/n$, y se tiene que la probabilidad de un evento $A$ es:

$$\mathbb{P}(A) = \frac{\text{casos favorables}}{\text{casos totales}} = \frac{|A|}{|\Omega|}, \text{ para toda } A \subset \Omega.$$


## Probabilidad Condicional

$\rightarrow$ **Probabilidad condicional**

**Def.** Sean $A, B$ eventos tal que $\mathbb{P}(B) > 0$, entonces definimos la probabilidad del evento $A$ sabiendo la ocurrencia del evento $B$ como sigue:

$$\mathbb{P}(A \mid B) := \frac{\mathbb{P}(A \cap B)}{\mathbb{P}(B)}$$



## Aplicación Práctica: Modelo Laplaciano y Probabilidad Condicional

**Planteamiento:** Se lanzan dos dados justos de 6 caras. 
* Queremos calcular la probabilidad de que la suma de las caras sea 8 (Evento $A$).
* Luego, queremos calcular la misma probabilidad, pero sabiendo de antemano que el primer dado cayó en 3 (Evento $B$).

**Definición Matemática:**
1. **Espacio Muestral ($\Omega$):** Todas las combinaciones posibles de dos dados. 
   $$\Omega = \{(i,j) : i, j \in \{1, 2, 3, 4, 5, 6\}\}$$
   La cardinalidad (casos totales) es $|\Omega| = 36$.
2. **Evento $A$ (Suma igual a 8):** $$A = \{(i,j) \in \Omega : i + j = 8\}$$
3. **Evento $B$ (Primer dado es 3):**
   $$B = \{(3,j) \in \Omega\}$$

**Fórmulas a utilizar:**
* **Probabilidad Laplaciana de $A$:** $\mathbb{P}(A) = \frac{|A|}{|\Omega|}$
* **Probabilidad Condicional de $A$ dado $B$:** $\mathbb{P}(A \mid B) = \frac{|A \cap B|}{|B|}$

```python
print("--- Modelo Laplaciano y Probabilidad Condicional ---")

# 1. Construimos el Espacio Muestral (Omega)
# Generamos todas las parejas (i, j) posibles usando dos ciclos for anidados
omega = [(i, j) for i in range(1, 7) for j in range(1, 7)]
total_omega = len(omega)
print(f"1. Tamaño del espacio muestral |Omega|: {total_omega} casos posibles.")

# 2. Definimos y contamos el Evento A (La suma es 8)
evento_A = [(i, j) for (i, j) in omega if i + j == 8]
casos_A = len(evento_A)
print(f"2. Evento A (Suma = 8): {evento_A}")
print(f"   Casos favorables |A|: {casos_A}")

# 3. Definimos y contamos el Evento B (El primer dado es 3)
evento_B = [(i, j) for (i, j) in omega if i == 3]
casos_B = len(evento_B)
print(f"3. Evento B (Primer dado = 3): {evento_B}")
print(f"   Casos posibles del nuevo escenario |B|: {casos_B}")

# 4. Encontramos la Intersección (A y B ocurren al mismo tiempo)
# (Es decir, la suma es 8 Y el primer dado es 3)
interseccion_AB = [resultado for resultado in evento_A if resultado in evento_B]
casos_interseccion = len(interseccion_AB)
print(f"4. Intersección (A n B): {interseccion_AB}")
print(f"   Casos favorables |A n B|: {casos_interseccion}")

print("\n--- Resultados Finales ---")

# 5. Cálculo de la Probabilidad Laplaciana (Clásica)
prob_A_laplace = casos_A / total_omega
print(f"Probabilidad de que la suma sea 8: P(A) = {casos_A}/{total_omega} = {prob_A_laplace:.4f}")

# 6. Cálculo de la Probabilidad Condicional
# Usamos la versión de cardinalidad de Laplace: |A n B| / |B|
prob_A_condicional = casos_interseccion / casos_B
print(f"Prob. de suma 8 DADO que el 1er dado es 3: P(A|B) = {casos_interseccion}/{casos_B} = {prob_A_condicional:.4f}")
```

# ***Modelo Frecuentista***

Consideremos el experimento de lanzar una moneda:

$\color{red}{\blacktriangleright}$ 1 lanzamiento: 1 cara. $\leftarrow$ No podemos decir nada.

$\color{red}{\blacktriangleright}$ 3 lanzamientos: 2 caras y 1 sol. $\leftarrow$ No podemos decir nada.

$\color{red}{\blacktriangleright}$ 300 lanzamientos: 200 caras y 100 soles. $\leftarrow$ Empezamos a hacernos preguntas.

$\color{red}{\blacktriangleright}$ 3000 lanzamientos: 2000 caras y 1000 soles. $\leftarrow$ Es casi seguro que la moneda está cargada.

La repetición del experimento es la que nos da la información, y tenemos que la frecuencia empírica del evento cara es $\frac{2000}{3000} = \frac{2}{3}$.

Esta frecuencia nos da una idea de la posibilidad de obtener cara en el lanzamiento de la moneda porque el número de repeticiones es grande.

### Aplicación Práctica: Lanzamiento de una moneda cargada

```python
#Importar librerias para simulación y Graficas
import numpy as np
import matplotlib.pyplot as plt

# Asigno una semilla para que los numeros aleatorios sean reproducibles
np.random.seed(42)

# Defino parametros del evento
Prob_cara = 2/3
Lanzamientos = 3000

# Simulo los 3000 lanzamientos
Result = np.random.choice([0, 1], size=Lanzamientos, p=[1 - Prob_cara, Prob_cara])

# Calcular la suma acumulada de las caras
Caras_acumuladas = np.cumsum(Result)

# Generamos un arreglo con el número de intentos [1, 2, 3, ..., 3000]
intentos = np.arange(1, Lanzamientos + 1)

# APLICACIÓN DE LA FÓRMULA FRECUENTISTA: 
# Probabilidad = Frecuencia del evento / Número de repeticiones
Prob_frecuentista = Caras_acumuladas / intentos

# 1. Imprimimos el reporte en texto para los hitos específicos
print("--- Resultados de la Simulación Empírica ---")
hitos = [1, 3, 300, 3000]
for n in hitos:
    # Restamos 1 porque los índices en Python empiezan en 0
    caras_obtenidas = Caras_acumuladas[n-1] 
    frecuencia = Prob_frecuentista[n-1]
    print(f"Lanzamientos: {n:<4} | Caras obtenidas: {caras_obtenidas:<4} | Frecuencia Empírica: {frecuencia:.4f}")

# 2. Generamos la visualización (Gráfica de convergencia)
plt.figure(figsize=(5, 2.5))

# Graficamos la línea de la simulación
plt.plot(intentos, Prob_frecuentista, label='Frecuencia Empírica (Simulada)', color='blue', alpha=0.8)

# Graficamos una línea roja horizontal que representa la teoría (2/3)
plt.axhline(y=Prob_cara, color='red', linestyle='--', label='Probabilidad Teórica (2/3)')

# Añadimos formato y etiquetas para cumplir con la documentación
plt.title("Modelo Frecuentista: Lanzamiento de Moneda Cargada")
plt.xlabel("Número de Lanzamientos")
plt.ylabel("Probabilidad (P(Cara))")
plt.ylim(0, 1) # Fijamos el eje Y de 0 a 1 porque es una probabilidad
plt.legend()
plt.grid(True, alpha=0.3)

# Mostramos la gráfica
plt.show()
```

# ***Modelo Geométrico***

En este modelo se busca encontrar probabilidades de la siguiente manera:

$$\mathbb{P}(A) = \frac{\text{area deseada}}{\text{área total}}$$

<span style="color:red;">Ejemplos:</span>

$\color{red}{\blacktriangleright}$ Problema del encuentro (Romeo y Julieta).

$\color{red}{\blacktriangleright}$ Aguja de Buffon.




### Aplicación Práctica: El Encuentro de Romeo y Julieta

**Planteamiento:** Romeo y Julieta acuerdan encontrarse en un lugar determinado entre las 12:00 y las 13:00 horas (un intervalo de 1 hora). Ambos llegan de forma independiente y aleatoria en ese lapso. El primero que llega espera exactamente 15 minutos (es decir, $1/4$ de hora) al otro. Si no llega, se va. ¿Cuál es la probabilidad de que se encuentren?

**Definición Geométrica:**
* Sean $x$ el tiempo de llegada de Romeo e $y$ el tiempo de llegada de Julieta. 
* El espacio muestral es un cuadrado de $1 \times 1$: $\Omega = \{(x,y) : 0 \leq x \leq 1, 0 \leq y \leq 1\}$
* El evento de encuentro $A$ ocurre si la diferencia entre sus tiempos de llegada es menor o igual a $1/4$ de hora ($0.25$): 
  $$A = \{(x,y) \in \Omega : |x - y| \leq 0.25\}$$

Usando el **Modelo Geométrico**, la probabilidad es el área de la región de encuentro dividida por el área total de $\Omega$.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Configuración del problema
N = 20000 # Lanzamos 10,000 "puntos" (simulaciones) para rellenar el área
espera = 0.25 # El tiempo de espera máximo es 1/4 de hora (15 minutos)

# Generamos los tiempos de llegada aleatorios para Romeo (x) y Julieta (y)
# np.random.uniform(0, 1) genera números aleatorios uniformes entre 0 y 1
x = np.random.uniform(0, 1, N)
y = np.random.uniform(0, 1, N)

# 2. Condición del evento (¿Se encuentran?)
# np.abs() calcula el valor absoluto. Se encuentran si |x - y| <= 0.25
encuentro = np.abs(x - y) <= espera

# 3. Cálculo de la probabilidad (Geométrica / Frecuentista)
casos_favorables = np.sum(encuentro) # Cuenta cuántos puntos cumplieron la condición
prob_simulada = casos_favorables / N
prob_teorica = 1 - (0.75 * 0.75) # Cálculo del área teórica (1 - el área de los dos triángulos de las esquinas)

print("--- Resultados del Problema del Encuentro ---")
print(f"Probabilidad Simulada por Python: {prob_simulada:.4f}")
print(f"Probabilidad Teórica (Área):      {prob_teorica:.4f} (equivale a 7/16)")

# 4. Visualización del Espacio Muestral y el Evento
plt.figure(figsize=(5, 5)) # Hacemos la gráfica cuadrada para representar Omega fielmente

# Dibujamos los puntos donde NO se encuentran (Rojo)
# La virgulilla (~) invierte la condición: selecciona los puntos que fueron "Falsos"
plt.scatter(x[~encuentro], y[~encuentro], color='crimson', alpha=0.2, s=5, label='No se encuentran')

# Dibujamos los puntos donde SÍ se encuentran (Verde)
plt.scatter(x[encuentro], y[encuentro], color='mediumseagreen', alpha=0.2, s=5, label='¡Se encuentran!')

# Dibujamos las líneas matemáticas que delimitan el área (y = x + 0.25 y y = x - 0.25)
linea_x = np.linspace(0, 1, 100)
plt.plot(linea_x, linea_x + espera, color='black', linestyle='--', label='Límite Matemático')
plt.plot(linea_x, linea_x - espera, color='black', linestyle='--')

# Detalles estéticos de la gráfica
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.title("Espacio Muestral $\Omega$: Encuentro de Romeo y Julieta")
plt.xlabel("Llegada de Romeo ($x$) en horas")
plt.ylabel("Llegada de Julieta ($y$) en horas")
plt.legend(loc='upper left')
plt.grid(True, alpha=0.2)

plt.show()
```

### Aplicación Práctica: La Aguja de Buffon

**Planteamiento:** Se deja caer aleatoriamente una aguja de longitud $L$ sobre una superficie plana que tiene líneas paralelas dibujadas, separadas por una distancia $D$ (donde $L \leq D$). ¿Cuál es la probabilidad de que la aguja toque o cruce una de las líneas?

**Definición Geométrica:**
Para modelar esto, necesitamos dos variables aleatorias:
1. Sea $y$ la distancia desde el centro de la aguja hasta la línea paralela más cercana. Como máximo, el centro puede estar exactamente a la mitad de dos líneas, por lo que: $0 \leq y \leq \frac{D}{2}$.
2. Sea $\theta$ el ángulo agudo que forma la aguja con las líneas paralelas: $0 \leq \theta \leq \frac{\pi}{2}$.

* **Espacio Muestral:** Es el rectángulo formado por todas las combinaciones posibles de posición y ángulo: 
  $$\Omega = \left\{(\theta, y) : 0 \leq \theta \leq \frac{\pi}{2}, 0 \leq y \leq \frac{D}{2}\right\}$$
* **Condición de cruce (Evento A):** Con un poco de trigonometría básica, sabemos que la aguja cruza una línea si la distancia $y$ es menor o igual al cateto opuesto formado por la mitad de la aguja: 
  $$A = \left\{(\theta, y) \in \Omega : y \leq \frac{L}{2} \sin(\theta)\right\}$$

La probabilidad teórica mediante integración geométrica es $\mathbb{P}(A) = \frac{2L}{\pi D}$.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Configuracion de la Aguja de Buffon
N = 10000     # Numero de agujas que vamos a lanzar
L = 1.0       # Longitud de la aguja
D = 1.0       # Distancia entre las lineas (Asumimos L = D para simplificar)

np.random.seed(42) # Semilla para reproducibilidad

# 2. Generamos las variables aleatorias (Lanzamiento)
# Angulo theta: distribucion uniforme entre 0 y pi/2 (90 grados)
theta = np.random.uniform(0, np.pi / 2, N)
# Distancia y: distribucion uniforme entre 0 y D/2
y = np.random.uniform(0, D / 2, N)

# 3. Evaluamos el evento (¿Cruza la linea?)
# Aplicamos la formula del Markdown: y <= (L/2) * sin(theta)
cruza = y <= (L / 2) * np.sin(theta)

# 4. Calculos de Probabilidad y estimacion de Pi
casos_favorables = np.sum(cruza)
prob_simulada = casos_favorables / N
prob_teorica = (2 * L) / (np.pi * D)

# Estimacion de Pi (Despejando Pi de la formula teorica usando la prob_simulada)
# Pi aprox = (2 * L) / (P_empirica * D)
pi_estimado = (2 * L) / (prob_simulada * D)

print("--- Resultados de la Aguja de Buffon ---")
print(f"Agujas cruzadas:      {casos_favorables} de {N}")
print(f"Probabilidad (Sim):   {prob_simulada:.4f}")
print(f"Probabilidad (Teor):  {prob_teorica:.4f}")
print(f"Estimacion de Pi:     {pi_estimado:.5f} (Valor real: {np.pi:.5f})")

# 5. Visualizacion del Modelo Geometrico
plt.figure(figsize=(6, 3))

# Dibujamos las agujas que NO cruzaron (Rojo)
plt.scatter(theta[~cruza], y[~cruza], color='crimson', alpha=0.3, s=5, label='No cruza')

# Dibujamos las agujas que SI cruzaron (Verde)
plt.scatter(theta[cruza], y[cruza], color='mediumseagreen', alpha=0.3, s=5, label='Cruza la linea')

# Dibujamos la curva matematica que separa ambos eventos: y = (L/2) * sin(theta)
linea_theta = np.linspace(0, np.pi / 2, 100)
curva_y = (L / 2) * np.sin(linea_theta)
plt.plot(linea_theta, curva_y, color='black', linewidth=2, label='Frontera: $y = \\frac{L}{2} \\sin(\\theta)$')

# Detalles esteticos
plt.title("Modelo Geometrico: Espacio Muestral de la Aguja de Buffon")
plt.xlabel("Angulo $\\theta$ (Radianes)")
plt.ylabel("Distancia $y$ a la linea mas cercana")
plt.xlim(0, np.pi / 2)
plt.ylim(0, D / 2)
plt.legend(loc='upper right')
plt.grid(True, alpha=0.3)

plt.show()
```
